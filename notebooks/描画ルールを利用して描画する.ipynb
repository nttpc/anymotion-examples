```python
from anymotion_sdk import Client
from IPython.display import Image, display_png, display_jpeg
```

```python
client = Client()
```

```python
!ls ../data
```

```python
display_png(Image("../data/squat.png"))
```

```python
keypoint_id = 12056
```

```python
drawing_id = client.draw_keypoint(
    keypoint_id,
    rule=[
        {"drawingType": "stickPicture", "pattern": "all", "color": "white", "size": 3},
        {"drawingType": "keypoint", "pattern": "all", "size": 5},
    ],
)
```

```python
drawing_id
```

```python
drawing_result = client.wait_for_drawing(drawing_id)
```

```python
drawing_result
```

```python
client.download(drawing_id, path="../data/squat_drawn_rule1.jpg")
```
